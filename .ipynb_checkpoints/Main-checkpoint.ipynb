# Movie Dataset Analysis and Ratings Prediction

* Download the 1 million ratings data set from the MovieDataSet.zip on Canvas
* Merge all of the data into a single Pandas DataFrame
* Store the data into an HDF5 file.

```python
import pandas as pd
#-------------------------------
#read in ratings data 
ratings = pd.read_csv("MovieDataSet/ratings.dat", sep="::", header=None, names = ["UserID", "MovieID", "Rating", "Timestamp"], engine="python")
print(ratings.head())

#read in and format users data 
users = pd.read_csv("MovieDataSet/users.dat", sep="::", header=None, names = ["UserID", "Gender", "Age", "Occupation","Zip-code"], engine="python")
age_mappings = {1:"under 18", 18:"18-24", 25:"25-34", 35:"35-44", 45:"45-49", 50:"50-55", 56:"56+"}
occupation_mappings={ 0:  "other" or not specified,
    1:  "academic/educator",
    2:  "artist",
    3:  "clerical/admin",
    4:  "college/grad student",
    5:  "customer service",
    6:  "doctor/health care",
    7:  "executive/managerial",
    8:  "farmer",
    9:  "homemaker",
    10:  "K-12 student",
    11:  "lawyer",
    12:  "programmer",
    13:  "retired",
    14:  "sales/marketing",
    15:  "scientist",
    16:  "self-employed",
    17:  "technician/engineer",
    18:  "tradesman/craftsman",
    19:  "unemployed",
    20:  "writer" }
#map age to age groups and occupation to occupation specification 
users["Age-Group"] = users["Age"].apply(lambda x: age_mappings[x])
users["Occupation-Group"] = users["Occupation"].apply(lambda x: occupation_mappings[x])
print(users.head())

#read in movies data 
movies = pd.read_csv("MovieDataSet/movies.dat", encoding='ISO-8859-1', sep="::", header=None, names = ["MovieID", "Title", "Genres"], engine="python")
    
print(movies.head())
```

```python
#for the movies data, organize the different genres in to a list and extract the movie year
# supress warning messages, not recommended but would make the final notebook submission cleaner
import warnings
warnings.filterwarnings("ignore")

movies['Genre-List'] = ""
movies['Movie-Year'] = ""
for index, row in movies.iterrows():
    movies['Genre-List'][index] = row['Genres'].split('|')
    movies['Movie-Year'][index] = int(row['Title'][-5:-1])
#movies = movies.drop('Genres', 1)
```

```python
print(movies.head())
```

```python
#merge datasets together using left join on Ratings (dropping rows that don't exist in both datasets for search key)
merged_df = ratings.merge(users, on="UserID").merge(movies, on="MovieID")
sorted_df = merged_df.sort_values(by="UserID").reset_index(drop=True) #sort dataframe by user ids
sorted_df
```

```python
# write file to hdf5 
sorted_df.to_hdf("mergedDataframe.h5", key="df", mode="w")
```

## Reporting Some Basic Details of the Data Collected.

*  How many movies have an average rating over 4.5 overall?
*  How many movies have an average rating over 4.5 among men?  How about women?
*  How many movies have an *median* rating over 4.5 among men over age 30?  How about women over age 30?
*  What are the ten most popular movies?
*  Make some conjectures about how easy various groups are to please? Support your answers with data!  

```python
# if you need to read in saved file, uncomment, 
# import pandas as pd
# sorted_df = pd.read_hdf("mergedDataframe.h5")
```

```python
#slicing sorted_df 
dfStats = sorted_df[['MovieID', 'Rating', 'Gender', 'Age', 'Age-Group', 'Occupation-Group', 'Genre-List', 'Movie-Year']]
#female data
dfFem = dfStats[dfStats['Gender']=='F']
#male data
dfMal = dfStats[dfStats['Gender']=='M']
```

```python
#find average ratings and merge
dfStatsAvg = (dfStats.groupby('MovieID')
            .Rating.mean()
            .rename('AvgRating')
            .reset_index()
            .sort_values(by='MovieID', ascending=False))

dfFemAvg = (dfFem.groupby('MovieID')
            .Rating.mean()
            .rename('AvgRating')
            .reset_index()
            .sort_values(by='MovieID', ascending=False))

# calculate males average rating for every movie
dfMalAvg = (dfMal.groupby('MovieID')
            .Rating.mean()
            .rename('AvgRating')
            .reset_index()
            .sort_values(by='MovieID', ascending=False))

# dfStatsAvgMerge = dfStatsAvg.merge(dfStats, on='MovieID')

#find average ratings over 4.5
dfTopAvg = dfStatsAvg[dfStatsAvg['AvgRating']>4.5]

#average rating over 4.5 among men
dfTopAvgMen = dfMalAvg[dfMalAvg['AvgRating']>4.5]

#average rating over 4.5 among women
dfTopAvgWomen = dfFemAvg[dfFemAvg['AvgRating']>4.5]

print ("Number of movies that have an average rating over 4.5 overall:", dfTopAvg.shape[0])
print ("Number of movies that have an average rating over 4.5 among men:", dfTopAvgMen.shape[0])
print ("Number of movies that have an average rating over 4.5 among women:", dfTopAvgWomen.shape[0])
```

```python
#find median ratings 
dfStatsMed = (dfStats.groupby('MovieID')
            .Rating.median()
            .rename('MedRating')
            .reset_index()
            .sort_values(by='MovieID', ascending=False))

#find median rating over 4.5
dfTopMed = dfStatsMed[dfStatsMed['MedRating']>4.5]

#median rating for all genders over 30
dfStatsOver30 = dfStats[dfStats['Age']>30]

dfStatsOver30Med = (dfStatsOver30.groupby('MovieID')
            .Rating.median()
            .rename('MedRating')
            .reset_index()
            .sort_values(by='MovieID', ascending=False))

dfTopMed30 = dfStatsOver30Med[dfStatsOver30Med['MedRating']>4.5]

#median rating over 4.5 for men over 30
dfMalOver30 = dfMal[dfMal['Age']>30]

dfMalOver30Med = (dfMalOver30.groupby('MovieID')
            .Rating.median()
            .rename('MedRating')
            .reset_index()
            .sort_values(by='MovieID', ascending=False))

dfTopMedMen = dfMalOver30Med[dfMalOver30Med['MedRating']>4.5]

#median rating over 4.5 for women over 30
dfFemOver30 = dfFem[dfFem['Age']>30]

dfFemOver30Med = (dfFemOver30.groupby('MovieID')
            .Rating.median()
            .rename('MedRating')
            .reset_index()
            .sort_values(by='MovieID', ascending=False))

dfTopMedWomen = dfFemOver30Med[dfFemOver30Med['MedRating']>4.5]

print ("Number of movies that have a median rating over 4.5 overall:", dfTopMed.shape[0])
print ("Number of movies that have a median rating over 4.5 among men over age 30:", dfTopMedMen.shape[0])
print ("Number of movies that have a median rating over 4.5 among women over age 30:", dfTopMedWomen.shape[0])
print ("Number of movies that have a median rating over 4.5 all genders over age 30:", dfTopMed30.shape[0])
```

```python
#calculate the number of the times each movie has been rated
dfStatsCnt = (dfStats.groupby('MovieID')
            .Rating.count()
            .rename('CntRating')
            .reset_index()
            .sort_values(by='CntRating', ascending=True))
```

```python
#Top 10 movies = movies with highest average rating and has been rated over 2000 times

dfPopular = (dfStatsCnt.merge(merged_df, on="MovieID")
             .merge(dfStatsAvg, on="MovieID")[["MovieID", "Title", "CntRating", "AvgRating"]]
             .drop_duplicates(subset="MovieID")
             .sort_values(by="CntRating", ascending=False)
             .sort_values(by="AvgRating", ascending=False).reset_index(drop=True))
dfPopular[(dfPopular['CntRating']>2000)&(dfPopular['AvgRating']>4)].head(10)
```

```python
# Conjecture: older users will likely to be harder to please because they've seen the most number of movies in their lives
# we hypothesize that the movies that have a high average and median rating of over 4.5 will decrease with increasing age group

# calculate median rating over a threshold for an age group
def calc_med_rating_high(age_group, high_rating_threshold, gender, printout=False):
    if gender=='F':
        df = dfStats[ (dfStats['Age-Group']==age_group) & (dfStats['Gender']=='F') ]
    elif gender=='M':
        df = dfStats[ (dfStats['Age-Group']==age_group) & (dfStats['Gender']=='M') ]
    else:
        df = dfStats[dfStats['Age-Group']==age_group]
        
    dfMedRating = (df.groupby('MovieID')
                   .Rating.median()
                   .rename('MedRating')
                   .reset_index()
                   .sort_values(by='MovieID', ascending=False))
    dfMedRatingHigh = dfMedRating[dfMedRating['MedRating'] > high_rating_threshold]
    
    if printout:
        print("There are "+str(df.shape[0]), gender, "user ratings in this age group: "+age_group)
        print("Number of movies that have a median rating over "+str(high_rating_threshold)+" is: "+str(dfMedRatingHigh.shape[0])+
              ", which is "+str(round(dfMedRatingHigh.shape[0]/dfMedRating.shape[0]*100, 2))+"% of all movie ratings in this age group.")
    
    return(dfMedRatingHigh)

dfStatsAge1 = calc_med_rating_high("under 18", 4.5, "total", printout=True)
dfStatsAge18 = calc_med_rating_high("18-24", 4.5, "total", printout=True)
dfStatsAge25 = calc_med_rating_high("25-34", 4.5, "total", printout=True)
dfStatsAge35 = calc_med_rating_high("35-44", 4.5, "total", printout=True)
dfStatsAge45 = calc_med_rating_high("45-49", 4.5, "total", printout=True)
dfStatsAge50 = calc_med_rating_high("50-55", 4.5, "total", printout=True)
dfStatsAge56 = calc_med_rating_high("56+", 4.5, "total", printout=True)
```

```python
# calculate average rating over a threshold for an age group
def calc_avg_rating_high(age_group, high_rating_threshold, gender, printout=False):
    if gender=='F':
        df = dfStats[ (dfStats['Age-Group']==age_group) & (dfStats['Gender']=='F') ]
    elif gender=='M':
        df = dfStats[ (dfStats['Age-Group']==age_group) & (dfStats['Gender']=='M') ]
    else:
        df = dfStats[dfStats['Age-Group']==age_group]
        
    dfAvgRating = (df.groupby('MovieID')
                   .Rating.mean()
                   .rename('AvgRating')
                   .reset_index()
                   .sort_values(by='MovieID', ascending=False))
    dfAvgRatingHigh = dfAvgRating[dfAvgRating['AvgRating'] > high_rating_threshold]
    if printout:
        print("There are "+str(df.shape[0]), gender, "user ratings in this age group: "+age_group)
        print("Number of movies that have an average rating over "+str(high_rating_threshold)+" is: "+str(dfAvgRatingHigh.shape[0])+
              ", which is "+str(round(dfAvgRatingHigh.shape[0]/dfAvgRating.shape[0]*100, 2))+"% of all movie ratings in this age group.")
    
    return(dfAvgRatingHigh)

dfStatsAge1 = calc_avg_rating_high("under 18", 4.5, "total", printout=True)
dfStatsAge18 = calc_avg_rating_high("18-24", 4.5, "total", printout=True)
dfStatsAge25 = calc_avg_rating_high("25-34", 4.5, "total", printout=True)
dfStatsAge35 = calc_avg_rating_high("35-44", 4.5, "total", printout=True)
dfStatsAge45 = calc_avg_rating_high("45-49", 4.5, "total", printout=True)
dfStatsAge50 = calc_avg_rating_high("50-55", 4.5, "total", printout=True)
dfStatsAge56 = calc_avg_rating_high("56+", 4.5, "total", printout=True)
```

# Correlation - Men versus Women

Let's look more closely at the relationship between the pieces of data we have.

* Make a scatter plot of men versus women and their mean rating for every movie.
* Make a scatter plot of men versus women and their mean rating for movies rated more than 200 times.
* Compute the *correlation coefficent* between the ratings of men and women.
* Conjecture under what circumstances the rating given by one gender can be used to predict the rating given by the other gender.

```python
# Compute the correlation coefficent between the ratings of men and women
def calc_pearson_corr_coeff(X_df, Y_df, X_label, Y_label, age_group):
    
    XY_merge = X_df.merge(Y_df, on='MovieID')
    #print(XY_merge)
    X = XY_merge['AvgRating_x']
    Y = XY_merge['AvgRating_y'] 
    
    print("=========================================================================")
    print("Age Group:", age_group)
    
    # sample size
    print('Number of movies that have both female and male ratings:', len(X)) 
    print('% of movies that have both female and male ratings:', str(round(len(X)/movies.shape[0]*100, 2))+"%") 
    
    # calculate sample mean
    meanX = X.mean()
    meanY = Y.mean()
    print('Mean of '+X_label+' is:', round(meanX, 3))
    print('Mean of '+Y_label+' is:', round(meanY, 3))
    
    # calculate standard deviation 
    stdX = X.std()
    stdY = Y.std()
    print('Sample standard deviation of '+X_label+' is:', round(stdX, 3))
    print('Sample standard deviation of '+Y_label+' is:', round(stdY, 3))
    
    # calculate sample covariance
    covXY = sum((X-meanX)*(Y-meanY))/len(X)
    print('Sample covariance is:', round(covXY, 3))
    
    # Pearson's correlation coefficient
    pearsonCoeff = covXY/(stdX*stdY)
    print('Pearson\'s correlation coefficient is:', round(pearsonCoeff, 3))
    
    return (pearsonCoeff)
```

```python
# Using the mean rating for every movie
FemMalCorr = calc_pearson_corr_coeff(dfFemAvg, dfMalAvg, 'female average rating', 'male average rating', "All Age Groups (All Movies)")
# Using the mean rating for movies that are rated over 200 times
FemMalCorrCntOver200 = calc_pearson_corr_coeff(dfFemAvgCntOver200, dfMalAvgCntOver200, 'female average rating', 'male average rating', "All Age Groups (Movies Rated Over 200 Times Overall)")
# When a movie has over 200 rating counts, the Pearson's correlation coefficient increases 
```

```python
#DataEncoding 
rating_mapping = {5:1, 4:1, 3:0, 2:0, 1:0}
df["RatingEncoding"] = df["Rating"].apply(lambda x: rating_mapping[x])
training_df = pd.concat([df, pd.get_dummies(df[["Gender", "Age-Group"]],drop_first=True)], axis=1)
training_df 
```

```python
#Model Prep 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
x = training_df.iloc[:, 12:].drop("RatingEncoding", axis=1).drop("Movie-Year", axis=1)
y = training_df["RatingEncoding"]
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.9, test_size=0.1, shuffle=True) 
```

```python
# Visualize the target variable of the training set to see if clases are balanced 
p = sns.countplot(y_train)
p.set_xticklabels(['Low Rating','High Rating'])
plt.show()
```

```python
#Train and test model using default parameters
clf = DecisionTreeClassifier(random_state = 0)
model = clf.fit(x_train, y_train)
preds = model.predict(x_test)
```

```python
#Model performance 
print("Model accuracy on test set:", accuracy_score(y_test, preds)) 
tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
print("Number of True negatives: ", tn)
print("Number of False positives ", fp)
print("Number of False negatives: ", fn)
print("Number of True positives: ", tp)
```
